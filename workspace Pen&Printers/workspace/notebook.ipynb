**Data Validation:**

The dataset contains 15000 rows and 8 columns before cleaning and validataion. I have validated all the columns against the criteria in the dataset table:

week: There were 6unique numeric values ranging from 1 to 6 that corresponds to the given description, there were no missing values, no changes were made to this column.

sales_method: There were 5unique character values: Email, Call, Email + Call, em + call and email before validation.The rows that had Email + Call and  em + call were changed to Email and Call, similarly the rows that had email were changed to Email. This left us with 3 unique values Email,Call,Email and Call that match the given description,there were no missing values.

customer_id: There were 15000 character type values that acts as unique identifiers for the customers that match the given description. There were no missing values. No changes were made to this column.

nb_sold: This column had 10unique values that are numeric and ranged from 7 to 16 that match the given description. There were no missing values, no changes were made to this column.

revenue: This column was numeric and was rounded to 2 decimal places that match the given description, however there were 1074 missing values. The missing values were replaced with the median value of 89.5 as the number of missing values were more than 5% of the total records.

years_as_customer: There were 42unique numeric values, however from the given description the company started in 1984 meaning that values in this column can not exceed 39 which is the number of years the company had existed, hence two values 63 and 47 were dropped, and the values left ranged from 0 to 39 that match the given description. There were no missing values.

nb_site_visits: This column had 27unique numeric values that match the given description.There were no missing values. No changes were made to this column.

state: The column had 50 unique states which is consistent with the number of states in the US. The values were character type. There were no missing values. No changes were made to this column.

After validation and cleaning the dataset had 14998 rows and 8 columns with no missing values


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

product_sales = pd.read_csv('https://s3.amazonaws.com/talent-assets.datacamp.com/product_sales.csv')
product_sales
```

```python
print(product_sales.info())
print(product_sales.isna().sum())
```

```python
product_sales['sales_method']=product_sales['sales_method'].str.replace('em ','Email ')
product_sales['sales_method']=product_sales['sales_method'].str.replace('+', 'and')
product_sales['sales_method']=product_sales['sales_method'].str.replace('email','Email')
product_sales['sales_method']=product_sales['sales_method'].str.replace('call','Call')
product_sales['sales_method'].value_counts()
product_sales['revenue'] = product_sales['revenue'].fillna(89.50)
product_sales = product_sales[product_sales['years_as_customer']<40]
product_sales.describe()
#product_sales.shape
```
